```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns


```

```python
data=pd.read_csv("used_car_dataset.csv") 
data
```

```python
brand_counts = data['Brand'].value_counts()


plt.figure(figsize=(10, 6))
brand_counts.plot(kind='bar', color='skyblue')
plt.title('Brand Popularity in Used Car Listings')
plt.xlabel('Brand')
plt.ylabel('Number of Cars Listed')
plt.xticks(rotation=90)
plt.show()
```

```python
import matplotlib.pyplot as plt

# Clean the 'AskPrice' column
data.AskPrice = data.AskPrice.str.replace("₹", '')
data.AskPrice = data.AskPrice.str.replace(",", '')
data.AskPrice = data.AskPrice.astype('int')

# Display the filtered data
display(data[data['AskPrice'] < 20000000])

# Generate the histogram
plt.figure(figsize=(10, 6))
plt.hist(data[data['AskPrice'] < 20000000]['AskPrice'], bins=10000, color='skyblue', edgecolor='black')
plt.title('Distribution of Asking Prices for Used Cars')
plt.ticklabel_format(style='plain')
plt.xlabel('Asking Price (INR)')
plt.ylabel('Number of Cars')
plt.show()
```

```python
import matplotlib.pyplot as plt

# Assuming 'data' is your DataFrame and 'Age' and 'AskPrice' are the columns of interest
avg_price_by_age = data.groupby('Age')['AskPrice'].mean()

plt.figure(figsize=(20, 16))
plt.hist(avg_price_by_age.values, bins=5, color='skyblue', edgecolor='black')
plt.title('Average Car Depreciation by Age')
plt.xlabel('Average Asking Price (INR)')
plt.ylabel('Frequency')
plt.grid(True)
plt.show()

```

```python
display(data[data["Age"]==25])
data.drop(data[data["Age"]==25].index,inplace=True)
display(data[data["Age"]==25])
```

```python
ownership_counts = data['Owner'].value_counts()


plt.figure(figsize=(10, 6))
ownership_counts.plot(kind='bar', color='skyblue')
plt.title('Ownership History of Used Cars')
plt.xlabel('Ownership Category')
plt.ylabel('Number of Cars')
plt.xticks(rotation=45)
plt.show()
```

```python
model_counts = pd.DataFrame(data['model'].value_counts())
model_counts=model_counts.iloc[:20,:]
# Plot the bar chart
plt.figure(figsize=(12, 8))
model_counts.plot(kind='bar', color='skyblue')
plt.title('Count of Sales for Individual Car Models')
plt.xlabel('Car Model')
plt.ylabel('Number of Sales')
plt.xticks(rotation=90)
plt.show()

```

# Based on the visualizations and graphs from the dataset, several key inferences can be drawn about the used car market:

1. Brand Popularity: Maruti Suzuki, Hyundai, and Honda are among the most popular brands, indicating their high trust and preference among consumers due to their reliability, affordability, and extensive service networks.

2. Price Range: The asking prices for used cars vary widely, with many cars priced between ₹2,00,000 and ₹6,00,000. This diversity caters to different budget segments, making the market accessible to a wide range of buyers.

3. Depreciation: Cars depreciate in value over time, with the steepest depreciation occurring in the first few years. The rate of depreciation slows down after the initial years.

4. Popular Models: Certain car models, such as Maruti Suzuki Swift, Hyundai i20, and Honda City, have significantly higher sales, indicating strong demand and popularity in the used car market.
